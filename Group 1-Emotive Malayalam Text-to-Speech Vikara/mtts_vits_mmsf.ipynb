```python
from transformers import VitsModel, AutoTokenizer
import torch

model = VitsModel.from_pretrained("facebook/mms-tts-mal")
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-mal")

text = "അവൾക്ക് ഞാൻ വാക്ക് കൊടുത്തതാണ്. നിഹാലിന്റെ ശബ്ദം ആ സ്വീകരണമുറിയിൽ പ്രതിധ്വനിച്ചിച്ചു "
inputs = tokenizer(text, return_tensors="pt")

with torch.no_grad():
    output = model(**inputs).waveform
```

```python
from IPython.display import Audio

Audio(output, rate=model.config.sampling_rate)
```
